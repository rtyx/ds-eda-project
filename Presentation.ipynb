## Methodology (Continued)

### Step 4: Hypothesis Testing
- Tested 3 hypotheses about affordable neighborhoods

### Step 5: Recommendations
- Identified top 3 investment properties
- Calculated profit margin potential
- Prioritized social impact areas


## Key Insight #1: Geographic Patterns

**Finding:** Price varies significantly by location

- **Affordable zipcodes** (bottom 25%): Average prices **$200K - $400K**
- **Expensive zipcodes** (top 25%): Average prices **$600K - $1M+**
- Clear **geographic clustering** of affordable vs. expensive areas

**Implication for Erin:**
Focus investment in **affordable zipcodes** where purchase prices are lower, leaving room for renovation costs and profit margins.


## Key Insight #2: Renovation Opportunity

**Finding:** Affordable neighborhoods have more unrenovated properties

- **Cheapest zipcodes**: **~85%** of properties are unrenovated
- **Expensive zipcodes**: **~75%** of properties are unrenovated
- **10 percentage point difference** = more renovation opportunities

**Implication for Erin:**
Affordable neighborhoods offer **more properties** that haven't been renovated yet, providing better opportunities to add value through improvements.


## Key Insight #3: Price Per Square Foot Advantage

**Finding:** Unrenovated properties offer better value

- **Unrenovated properties** in affordable areas: **~$89/sqft**
- **Renovated properties** in same areas: **~$120/sqft**
- **$31/sqft difference** = renovation adds significant value

**Implication for Erin:**
Buying unrenovated properties at lower prices per square foot creates **room for renovation costs** while maintaining profitability when selling at renovated prices.


## Key Insight #4: Profit Margin Reality Check

**Finding:** Most renovation projects show negative profit margins

- **Small houses** (< 2000 sqft): Average profit margin **-33.4%**
- **Large houses** (≥ 2000 sqft): Average profit margin **-25.0%**
- Only **0.8%** of small houses and **1.9%** of large houses achieve target margins (5-15%)

**Analysis:** Our hypothesis that smaller houses offer better profit margins was **NOT supported**. Both categories show negative margins, indicating renovation costs exceed value appreciation in most affordable zipcodes.

**Critical Finding:** The renovation strategy may not be profitable for most properties under current assumptions. However, we identified specific properties (top 3 recommendations) that do show positive margins.

**Implication for Erin:**
Be highly selective - focus on properties with verified positive profit potential rather than assuming size alone determines profitability.


## Recommendation #1: Target Affordable Zipcodes

**Action:** Focus investment in **bottom 25% zipcodes** by average price

**Why:**
- Lower purchase prices ($200K-$400K range)
- More unrenovated properties available
- Better price per square foot ratios
- Higher social impact potential

**Top Target Zipcodes:**
- Zipcode 98198 (and similar affordable areas)
- These areas have the highest concentration of investment opportunities


## Recommendation #2: Focus on Unrenovated Properties

**Action:** Prioritize properties with **yr_renovated = 0**

**Why:**
- **$31/sqft price difference** vs. renovated properties
- Clear value-add opportunity through renovation
- Lower purchase price = room for renovation costs
- Higher profit margin potential

**Selection Criteria:**
- Unrenovated (yr_renovated = 0)

- Manageable size (1000-2500 sqft)


## Recommendation #3: Be Highly Selective

**Action:** Focus on **specific properties** with verified positive profit margins, not size categories

**Why:**
- Our analysis shows most properties have **negative profit margins** after renovation
- Property size alone doesn't determine profitability
- The top 3 recommended properties (all 1,900-2,300 sqft) show **30%+ margins**
- These exceptions prove the strategy can work with careful selection

**Key Criteria:**
- Properties in affordable zipcodes (98198)
- Unrenovated with good condition (≥3)
- Specific price-per-sqft ratios that allow for profitable renovation
- Verified through our profit margin calculations

**Avoid:**
- Assuming any size category is automatically profitable
- Properties without verified positive profit margins
